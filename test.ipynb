```python
df=pd.read_csv("./sdrpy/data/sample_data.csv")
df
```

```python
from sdrpy.filters.filter_functions import get_trades 
df1 =  get_trades (df, product="xccy", product_type="Basis", currencies="CAD", maturity="m>3", date_range="-3w")
df1
```

```python
df1["Effective Date"]
```

```python
df1 = df1.loc[:, ~df1.columns.duplicated()]
```

```python
from datetime import date
import pandas as pd

today = date.today()
df1["Date"] = pd.to_datetime(df1["Date"])

# Convert today's date to Timestamp
today_timestamp = pd.Timestamp(today)

# Calculate the difference
diff = today_timestamp - df1["Date"]
df1["Day_diff"]=today_timestamp - df1["Date"]
# Now diff will contain the difference between today's date and each date in the DataFrame
```

```python
df1=df1[df1["Day_diff"]>=pd.Timedelta(days=20)]
df1
```
